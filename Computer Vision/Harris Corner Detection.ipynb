# Harris Corner Detection

Load and display the image

```python
import cv2
import numpy as np
import matplotlib.image as mpimg
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
image_color = cv2.imread('Chess Board.jpg')
cv2.imshow('Chess Board', image_color)
cv2.waitKey()
cv2.destroyAllWindows()
```

Convert to Grayscale and perform Harris Corner Detection

```python
image_gray = cv2.cvtColor(image_color, cv2.COLOR_BGR2GRAY)

corners = cv2.cornerHarris(image_gray, 5, 3, 0.1) # cv2.cornerHarris(img, block size, ksize, k)
plt.imshow(corners, cmap = 'gray')
plt.show()
```

```python
corners_dilate = cv2.dilate(corners, np.ones((8,8),np.uint8), iterations = 1)
plt.imshow(corners_dilate, cmap = 'gray')
```

```python
image_color[corners_dilate > 0.01 * corners_dilate.max() ] = [0, 255, 0]

cv2.imshow('Image with Corners', image_color)
cv2.waitKey()
cv2.destroyAllWindows()
```

Detect corners of a truck

```python
image_color = cv2.imread('Truck_2.png')

cv2.imshow('Find Truck Corners', image_color)
cv2.waitKey()
cv2.destroyAllWindows()
```

Convert to Grayscale

```python
image_gray = cv2.cvtColor(image_color, cv2.COLOR_BGR2GRAY)
plt.imshow(image_gray, cmap = 'gray')
```

```python
image_gray.shape
```

Perform Harris Corner Detection

```python
corners = cv2.cornerHarris(image_gray, 7, 5, 0.1)
corners_dilate = cv2.dilate(corners, np.ones((8,8),np.uint8), iterations = 1)
```

```python
plt.imshow(corners, cmap = 'gray')
plt.show()
```

```python
plt.imshow(corners_dilate, cmap = 'gray')
```

```python
# If pixel value is greater than a threshold, return true and mark it as green
Image_color_corners = image_color.copy()
Image_color_corners[corners_dilate > 0.03 * corners_dilate.max() ] = [0, 255, 0]

cv2.imshow('Image with Corners', Image_color_corners)
cv2.waitKey()
cv2.destroyAllWindows()
```
