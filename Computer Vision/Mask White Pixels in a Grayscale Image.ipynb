# Mask White Pixels in a Grayscale Image

Import all the necessary libraries

```python
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

Read and display the image

```python
image_color = mpimg.imread('Road_Lane_Colored.jpg')
image_gray = cv2.cvtColor(image_color, cv2.COLOR_BGR2GRAY)
plt.imshow(image_gray, cmap = 'gray')
image_gray.shape
```

Sort out the white color out of the image

```python
image_copy = np.copy(image_gray)
image_copy.shape
```

```python
image_copy[ (image_copy[:,:] < 250) ] = 0  # any value that is not white colour
```

Display the image 

```python
plt.imshow(image_copy, cmap = 'gray')
plt.show()
```
